```python
import pandas as pd

# Step 1: Load the original CSV file
input_file = '/content/faculty.csv'  # Replace with your file path
df = pd.read_csv(input_file)

# Step 2: Remove the unnecessary rows
df.drop(index=[0, 1], inplace=True)  # Drop rows with index 0 and 1

# Step 3: Extract specific columns
columns_to_extract = ['Q1', 'Q9#1_1', 'Q9#1_2', 'Q9#1_3', 'Q9#1_4',
                      'Q9#1_5', 'Q9#1_6', 'Q9#1_7', 'Q9#1_8', 'Q9#1_9',
                      'Q9#1_10', 'Q9#1_11', 'Q9#1_12']  # Replace with desired columns
extracted_df = df[columns_to_extract]

# Step 4: Rename 'Q1' to 'Full Name' for clarity
extracted_df.rename(columns={'Q1': 'Full Name'}, inplace=True)
extracted_df.rename(columns={'Q9#1_1': 'Technical contribution to field'}, inplace=True)
extracted_df.rename(columns={'Q9#1_2': 'Ability to explain complex ideas'}, inplace=True)
extracted_df.rename(columns={'Q9#1_3': 'Dependability, Reliability'}, inplace=True)
extracted_df.rename(columns={'Q9#1_4': 'Initiative'}, inplace=True)
extracted_df.rename(columns={'Q9#1_5': 'Interpersonal skills/Collaboration with others'}, inplace=True)
extracted_df.rename(columns={'Q9#1_6': 'Time Management'}, inplace=True)
extracted_df.rename(columns={'Q9#1_7': 'Work Ethics'}, inplace=True)
extracted_df.rename(columns={'Q9#1_8': 'Response to feedback'}, inplace=True)
extracted_df.rename(columns={'Q9#1_9': 'Listening to others'}, inplace=True)
extracted_df.rename(columns={'Q9#1_10': 'Professionalism'}, inplace=True)
extracted_df.rename(columns={'Q9#1_11': 'Willingness to have responsibility'}, inplace=True)
extracted_df.rename(columns={'Q9#1_12': 'Ability to follow instructions'}, inplace=True)




# Step 5: Map ratings to numerical values
rating_to_score = {
    'Poor': 1,
    'Average': 2,
    'Good': 3,
    'Very Good': 4,
    'Excellent': 5
}

# Apply mapping to all columns except 'Full Name'
for column in extracted_df.columns[1:]:
    extracted_df[column] = extracted_df[column].map(rating_to_score)

# Step 6: Add a column with the sum of rows
extracted_df['Total Score'] = extracted_df.iloc[:, 1:].sum(axis=1)  # Compute row-wise sum

# Step 7: Save the processed data to a new CSV file
output_file = "processed_file_with_row_sums.csv"
extracted_df.to_csv(output_file, index=False)

print(f"Processed file saved as {output_file}")
```
